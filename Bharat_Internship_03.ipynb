## **Data Preprocessing**

```python
import pandas as pd


data = pd.read_csv('/content/attrition_data.csv')
data['Attrition'] = data['Attrition'].map({'Yes': 1, 'No': 0})
factors_of_interest = ['DistanceFromHome', 'JobRole', 'JobSatisfaction', 'MonthlyIncome', 'OverTime']
```

## **Descriptive Statistics**

```python
descriptive_stats = data[factors_of_interest].describe()
print(descriptive_stats)
```

## ** Data Visualization**

```python
import matplotlib.pyplot as plt
import seaborn as sns

plt.figure(figsize=(8, 6))
sns.boxplot(x='Attrition', y='DistanceFromHome', data=data)
plt.xlabel('Attrition')
plt.ylabel('Distance From Home')
plt.title('Attrition vs. Distance From Home')
plt.show()

plt.figure(figsize=(12, 6))
sns.countplot(x='JobRole', hue='Attrition', data=data)
plt.xlabel('Job Role')
plt.ylabel('Count')
plt.title('Attrition by Job Role')
plt.legend(title='Attrition', labels=['No', 'Yes'], loc='upper right')
plt.xticks(rotation=45, ha='right')
plt.show()

plt.figure(figsize=(8, 6))
sns.countplot(x='JobSatisfaction', hue='Attrition', data=data)
plt.xlabel('Job Satisfaction')
plt.ylabel('Count')
plt.title('Attrition by Job Satisfaction')
plt.legend(title='Attrition', labels=['No', 'Yes'], loc='upper right')
plt.show()

plt.figure(figsize=(6, 4))
sns.countplot(x='OverTime', hue='Attrition', data=data)
plt.xlabel('OverTime')
plt.ylabel('Count')
plt.title('Attrition by OverTime')
plt.legend(title='Attrition', labels=['No', 'Yes'], loc='upper right')
plt.show()
```
